```python
#feature selection
# X = df.loc[:,'Gender':'Property_Area']
Y = df.loc[:,'Loan_Status']

X = df.loc[:,'Dependents':'Property_Area']

# X_test = test_df.loc[:,'Gender':'Property_Area']
X_test = test_df.loc[:,'Dependents':'Property_Area']
Y.replace({'Y':1,'N':0}, inplace=True)
Y.head()
```

```python
# X = pd.get_dummies(X, columns=["Gender", "Married","Education",'Self_Employed','Property_Area'])
# X_test = pd.get_dummies(X_test, columns=["Gender", "Married","Education",'Self_Employed','Property_Area'])

X = pd.get_dummies(X, columns=["Education",'Self_Employed','Property_Area'])
X_test = pd.get_dummies(X_test, columns=["Education",'Self_Employed','Property_Area'])
X.head()
```

```python
# #K fold
N_FOLDS = 8
max_iter = N_FOLDS
folds = StratifiedKFold(n_splits = max_iter)
y_pred = np.zeros(len(X_test))
features = X.columns

for fold_, (trn_idx, val_idx) in enumerate(folds.split(X, Y)):
    
    X_trn, y_trn = X.iloc[trn_idx][features], Y.iloc[trn_idx]
    X_val, y_val = X.iloc[val_idx][features], Y.iloc[val_idx]
    X_tst = X_test[features]
    
    clf = DecisionTreeClassifier(criterion="entropy", max_depth=4)
    clf = clf.fit(X_trn,y_trn)
    y_pred += clf.predict(X_test)/max_iter
    
y_pred = np.around(y_pred,0)
```

```python
result_df = pd.DataFrame()
result_df['Loan_ID'] = test_df['Loan_ID']
result_df['Loan_Status'] = y_pred
result_df['Loan_Status'].replace({1:'Y',0:'N'},inplace=True)
result_df.head()
result_df.set_index('Loan_ID')
result_df['Loan_Status'].value_counts()
```
